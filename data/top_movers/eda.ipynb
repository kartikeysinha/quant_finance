# Getting the data

```python
url = 'https://thestockmarketwatch.com/markets/pre-market/today.aspx'
```

```python
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/92.0.4515.159 Safari/537.36"
}
response = requests.get(url, headers=headers)
```

```python
soup = BeautifulSoup(response.content, "html.parser")
```

```python
gainers_table_header = soup.find("div", id="tdGainersDesktop")
losers_table_header = soup.find("div", id="tdLosersDesktop")

gainers_table = gainers_table_header.find("table", class_="tbldata")
losers_table = losers_table_header.find("table", class_="tbldata")
```

```python
def get_df_from_table_html(table):
    data = []
    headers = [header.text for header in table.find_all("th")]
    for row in table.find_all("tr")[1:]:
        cells = row.find_all("td")
        data.append([cell.text for cell in cells])

    # Create DataFrame
    df = pd.DataFrame(data, columns=headers)
    
    return df

df_gainers = get_df_from_table_html(gainers_table)
df_losers = get_df_from_table_html(losers_table)
```
